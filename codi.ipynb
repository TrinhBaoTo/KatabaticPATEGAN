```python
# Preprocess data
from pathlib import Path

dataset_path = ROOT / "raw_data" / "car.csv"
output_path = ROOT / "discretized_data" / "car.csv"
output_path.parent.mkdir(parents=True, exist_ok=True)

discretize_preprocess(str(dataset_path), str(output_path))

```

```python
from katabatic.models.codi.models import CODI
from katabatic.pipeline.train_test_split.pipeline import TrainTestSplitPipeline
# Run pipeline
input_csv = str(output_path)
output_dir = str(ROOT / "sample_data" / "car")
real_test_dir = output_dir

synthetic_dir = str(ROOT / "synthetic" / "car" / "codi")

# Use lighter training settings and force CPU to avoid kernel crashes
pipeline = TrainTestSplitPipeline(model=lambda: CODI(epochs=10, batch_size=256, device='cpu'))
result = pipeline.run(
    input_csv=input_csv,
    output_dir=output_dir,
    synthetic_dir=synthetic_dir,
    real_test_dir=real_test_dir,
)
print(result)

```
